What percentage of businesses have a rating greater than or 4.5?

```python
def percentages(per):
    l_counter = 0
    for l in pet_stores['rating']:
        if l >= per:
            l_counter += 1
    print((l_counter/len(pet_stores['rating']) * 100))

       
```

```python
# Collects the business_ids from the df
business_ids = []
mycsv = csv.reader(open('./petstores.csv'))
for row in mycsv:
    bus_id = row[1]
    business_ids.append(bus_id)
# stores the correct business ids    
business_ids = business_ids[1:]
```

```python
# Creates the business urls for the api
business_urls = []
for business in business_ids:
    business_url = 'https://api.yelp.com/v3/businesses/' + business + '/reviews'
    business_urls.append(business_url)
```

```python
reviews = yelp_call_reviews(creds, business_urls)
```
